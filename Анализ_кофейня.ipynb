Напишем функцию,которая распаковывает наименование продукта и его категорию с помощью модуля json

```python
def unpack_product(row):
  d = json.loads(row)
  return pd.Series({
      'product_name': d['product_name'],
      'product_category': d['product_category']
  })
```

```python
#Применяем функцию и добавляем новые столбцы
df_copy[['product_name','product_category']] = df_copy['product_info'].apply(unpack_product)
```

```python
#удаляем исходный словарный столбец
df_copy = df_copy.drop(columns = ['product_info'])
```

```python
df_copy.head(2)
```

###'store_info'

Напишем функцию,которая распаковывает название города с помощью модуля json

```python
def unpack_city(row):
  d = json.loads(row)
  return pd.Series({
      'store_city': d['store_city']
  })
```

```python
#Применяем функцию и добавляем новый столбец
df_copy[['store_city']] = df_copy['store_info'].apply(unpack_city)
```

```python
#удаляем исходный словарный столбец
df_copy = df_copy.drop(columns = ['store_info'])
```

Напишем функцию,которая распаковывает дату рождения и пол с помощью модуля json

```python
def unpack_customer(row):
  d = json.loads(row)
  return pd.Series({
      'birthdate': d['birthdate'],
      'gender': d['gender']
  })
```

```python
#Применяем функцию и добавляем новый столбец
df_copy[['birthdate','gender']] = df_copy['customer_info'].apply(unpack_customer)
```

## Расчет дополнительных признаков

### 'age'

 для расчета возраста необходимо из текущей даты вычесть дату рождения (признак "birthdate")

```python
# текущую дату можно получить с помощью метода Timestamp.today()
current_day = pd.Timestamp.today()
current_day
```

```python
# извлечем значение года из даты с помощью атрибута year
current_year = current_day.year
current_year
```

```python
# рассчитаем значения возраста и сохраним полученные значения в признак age
# при этом необходимо каждую дату рождения перевести в значение года аналогично через атрибут year
df_copy['age'] = current_year - df_copy.birthdate.apply(lambda x: x.year)
df_copy.age
```

```python
# исключим исходный столбец birthdate
df_copy.drop(columns='birthdate', inplace=True)
```
